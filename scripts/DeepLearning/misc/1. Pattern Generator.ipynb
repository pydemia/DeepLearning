# Pattern Augmentator (Real Size)

```python
from pydemia.manufacture.Lot import wafer as line_wafer
from pydemia.manufacture.new_lot import wafer as arc_wafer
import numpy as np
import scipy as sp
import cv2
import skimage as ski
from skimage import filters, transform
import matplotlib.pyplot as plt
import itertools as it
import os
from glob import glob
import pickle

from skimage.measure import ransac, CircleModel, LineModelND
from skimage.restoration import denoise_tv_chambolle, denoise_bilateral
```

```python
%matplotlib inline
```

# The Bad

## Y : .90

### Generate the Bad

```python
#%% wafer

# Create a wafer
alot_90 = arc_wafer(fac_id='Factory1', lot_cd='ABC', end_tm='20161101123456',
                    size=(40, 30), unit_cnt=30, f_type='arc', y_val=.9)
```

```python
alot = alot_90
for _ in alot.lotData['unit_id'].unique():
    alot.unitplot(columns=['xcoord', 'ycoord'],
                  lot_id='', unit_id=_,
                  val='val', figsize=(5, 4.7),
                  pattern=['B'], m=['s'], s=[50], c=['red'],
                  a=[.5], shown=True)
```

### Dump the Bad

```python
alot = alot_90
dirname = 'wafer_data/bad_base'
os.makedirs(dirname, exist_ok=True)
with open(dirname + '/' + 'bad_wafer_y90_base.dump', 'wb') as bak:
    pickle.dump(alot, bak)
```
